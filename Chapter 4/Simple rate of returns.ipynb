## Simple rate of returns


$$
\frac{End Price - Beginning Price}{Beginning Price} = \frac{End Price}{Beginning Price} - 1
$$

```python
MSFT['simple_return'] = (MSFT['Close'] / MSFT['Close'].shift(1)) - 1
print (MSFT['simple_return'])
```
